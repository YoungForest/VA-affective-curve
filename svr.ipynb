```python
import pandas as pd
arousalFeatures = features[1]
arousalFeaturesDf = pd.read_csv(arousalFeatures, index_col='id', sep='\t')
arousalFeaturesDf.head()
```

```python
ranking = pd.read_csv('/root/yangsen-data/LIRIS-ACCEDE-annotations/LIRIS-ACCEDE-annotations/annotations/ACCEDEranking.txt', sep='\t', index_col='id')
ranking.head()
```

```python
setsPd = pd.read_csv('/root/yangsen-data/LIRIS-ACCEDE-annotations/LIRIS-ACCEDE-annotations/annotations/ACCEDEsets.txt', sep='\t', index_col='id')
setsPd.head()
```

```python
withSetSplitLabel = pd.merge(pd.merge(ranking, setsPd, on='name'), arousalFeaturesDf, on='name')
withSetSplitLabel.head()
```

```python
trainSet = withSetSplitLabel[withSetSplitLabel['set'] >= 1]
print(trainSet.head())
len(trainSet)
```

```python
testSet = withSetSplitLabel[withSetSplitLabel['set'] == 0]
len(testSet)
```

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np
aouralFeatures = ['colorfulness', 'minEnergy', 'alpha', 'lightning', 'globalActivity', 'nbWhiteFrames', 'nbSceneCuts', 'cutLength', 'flatness_env', 'wtf_max2stdratio_1', 'wtf_max2stdratio_2', 'wtf_max2stdratio_3', 'wtf_max2stdratio_4', 'wtf_max2stdratio_5', 'wtf_max2stdratio_6', 'wtf_max2stdratio_7', 'wtf_max2stdratio_8', 'wtf_max2stdratio_9', 'wtf_max2stdratio_10', 'wtf_max2stdratio_11', 'wtf_max2stdratio_12', 'medianLightness', 'slope']
y = trainSet['arousalValue']
X = trainSet[aouralFeatures]
regr = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2))
regr.fit(X, y)
```

```python
Ytest = testSet['arousalValue']
Xtest = testSet[aouralFeatures]
Ypredict = regr.predict(Xtest)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Ytest, Ypredict)
```
